# Fine-Tuning
## Optimisation des hyperparamètres
Les principales stratégies sont :
1. la recherche dans une grille (**grid search**) : pour chaque hyperparamètre, un nombre donné de valeurs est fourni, et la recherche essaiera toutes les combinaisons possibles ;
2. la recherche aléatoire (**random search**) : seul un sous-ensemble des combinaisons possibles sera testé. La recherche aléatoire peut prendre ses valeurs parmi des listes fournies ou en les sélectionnant sur des intervalles continus ;
3. la recherche par gradient (**gradient-based optimization**) : en fonction des résultats obtenus sur différents ensembles, la recherche sera guidée dans le sens du gradient vers l’optimisation de la métrique choisie.

**Seules les deux premières existent dans Scikit-learn**. Il faut alors fournir à la méthode l’estimator utilisé, les hyperparamètres avec les valeurs ou les distributions, la ou les métriques à optimiser.



```python
import scipy.stats
# Recherche par grille - IRIS
params_grille = {'max_depth':[3, 5, 7],
         'splitter':['best', 'random'],
         'min_samples_split':[2, 5]}


# Recherche Aléatoire

params_aleatoire = {'max_depth': scipy.stats.randint(3, 8), # 8 non compris
         'splitter': ['best', 'random'],
         'min_samples_split': [2, 5]}
```

il faut ensuite créer l’estimateur puis la recherche. Celle-ci applique de la validation croisée. Enfin, il faut appeler la méthode fit de la recherche.

```python
# Grille
from sklearn.model_selection import GridSearchCV

tree_classifier = DecisionTreeClassifier(random_state=42)
grid_classifier = GridSearchCV(tree_classifier,
param_grid=params_grille, scoring='accuracy', cv=5)

grid_classifier.fit(train_X_iris, train_y_iris)



```

Une fois l’apprentissage fait, il est possible d’obtenir différentes informations sur la recherche exécutée, comme les scores des différentes possibilités testées.

Les deux attributs les plus utiles sont **best_estimator_** qui permet de récupérer le meilleur modèle et le set de paramètres utilisé, et **best_score_** qui fournit le score associé.

# Sur- et sous-apprentissage

Si un modèle est très bon sur les données d’apprentissage mais donne de mauvais résultats sur les données de validation, il fait alors preuve de sur-apprentissage (over-fitting en anglais).

Si un modèle, au contraire, n’a de bons résultats ni sur l’ensemble d’apprentissage ni sur l’ensemble de validation, c’est le signe qu’il n’y a pas d’apprentissage. Il s’agit alors de sous-apprentissage.



Cet équilibre entre ne pas apprendre et trop apprendre est aussi appelé **le compromis biais-variance**.
-L’erreur sur l’ensemble d’apprentissage est nommée le biais.
-L’erreur située entre l’ensemble d’apprentissage et l’ensemble de validation est notée la variance.

Généralement, sur un modèle, **plus le biais diminue et plus la variance augmente**, d’où la notion de compromis.




# Methodes ensemblistes

## Bagging
Le terme bagging vient de « Boostrap Aggregating ». Il consiste à agréger des modèles, chacun ayant été créé sur un sous-ensemble des données.



## Boosting
Le boosting est semblable au bagging, sauf que l'idée est de corriger les erreurs. Au lieu que les données soient choisient de manière aléatoire, À chaque étape on entraine un nouveau modèle qui se concentre sur ce qui a été mal prédit par le précédent.



## Stacking
Le stacking consiste à utiliser des modèles provenant d’algorithmes différents. On peut donc combiner les forces de plusieurs algos.

Le stacking permet de créer des couches de modèles (deux en général)
